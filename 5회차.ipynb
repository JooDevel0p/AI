```python
import numpy as np
import matplotlib.pyplot as plt

x=np.array([2,4,6,8])   
y=np.array([65,85,70,95])

a=0 
b=0
lr=0.005


for i in range(5000): 
  y_hat=a*x+b
  error=y_hat-y #오차
  a_diff=sum(2*error*x)     
  b_diff=sum(2*error*1)     

  a=a-a_diff*lr
  b=b-b_diff*lr  
  print(f'{i+1}회 학슴 // 기울기 {a} //절편 {b}')

plt.scatter(x,y)
plt.plot(x,a*x+b,'r') 
plt.show()  

#ㅇ예측이 안맞는 이유는 요인이 하나...
```

```python

import numpy as np
import matplotlib.pyplot as plt

x1=np.array([2,4,6,8])   #공부시간 
x2=np.array([1,2,1,3]) #문제집 푼 개수       #데이터 늘리기
y=np.array([65,85,70,95])

a1=0
a2=0   #기울기가 2개 생겼으니까... 
b=0
lr=0.005


for i in range(5000): 
  y_hat=a1*x1+a2*x2+b
  error=y_hat-y #오차
  a1_diff=sum(2*error*x1)
  a2_diff=sum(2*error*x2)         
  b_diff=sum(2*error*1)     

  a1=a1-a1_diff*lr
  a2=a2-a2_diff*lr
  b=b-b_diff*lr  

  print(f'{i+1}회 학습 // 기울기1 : {a1} //기울기2 : {a2}//절편 : {b}')

plt.scatter(x1,y)
plt.plot(x1, a1*x1+a2*x2+b, 'r')
plt.show()
```

```python
#도미, 빙어
#K-최근접 이웃 분류 모델


#도미 데이터(길이와 무게)-이 데이터로 학습을 시킬 예정
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
```

```python
plt.scatter(bream_length,bream_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
#빙어 데이터
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
plt.scatter(bream_length,bream_weight)
plt.scatter(smelt_length,smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
##학습에 들어가기 전에...머신러닝에 데이터를 넣어줄때는 머시러닝이 좋아하느 데이터 모야잉 있다.
##머신러닝의 좋은점 문제지와 정답지를 주면 알아서 학습하는 것이 좋은점
# 컴퓨터가 좋아하는 문제지와 정답지가 있따.
#문제지 : 2차원 넘파이 배열
#정답지 : 1차원 넘파이 배열   로 만들어 줘야 함..

length=bream_length+smelt_length
weight=bream_weight+smelt_weight
```

```python
print(length)  #문제지
```

```python
print(weight) #문제지
```

```python
fish_data=[]

for i,j in zip(length,weight):  #길이와 무게를 하나로 묶어서 fish_data에 넣기  #2차원 리스트 #이런 모양으로 머신러닝에 넣어야..
  fish_data.append([i,j])

fish_data
```

```python
#정답지

fish_target=[1]*35+[0]*14   #도미는 35,빙어 14
print(fish_target)
```

```python
##가장 가까운 데이터 5개를 찾는다!!도미 5개가 나왔네!!그럼 도미네!!
##빙어 2,도미3 그럼 도미!!

from sklearn.neighbors import KNeighborsClassifier

kn=KNeighborsClassifier()  #kn이란 변수는 클래스를 받았으니까 모든 라이브러리 사용가능
kn.fit(fish_data,fish_target)  #fit 이란 함수에 문제지와 답안지 넣어주기만 하면 알아서 학습
```

```python
#이 모델이 얼마나 정확한깋 알아보기  #왜 100점이야???학습한걸 넣었으니 당연...
kn.score(fish_data, fish_target)
```

```python
# 30,600 이라면 빙어일까 도미일까>??도미
kn.predict([[30,600]])  #2찬원 넘파이로 넣어야함..
```

```python
#컴퓨터가 모르는 데이터를 넣어도 잘 예측을 할 수 있는지 알아보자!!
#가장 가까운 데이터를 5개를 49개로 바꾼다면????

kn49=KNeighborsClassifier(n_neighbors=49)
kn49.fit(fish_data,fish_target)
```

```python
kn49.score(fish_data,fish_target) #정답룰떨어짐.. 왜냐면 빙어는 거의다 틀렸을 것..왜냐..데이터가 도미가 너 많구 다수결의 법칙이기기 어렵다..
```

```python
kn=KNeighborsClassifier()
kn.fit(fish_data,fish_target)

for n in range(1,50):    #개수를 올려가며 정확도 파악하기,18개 되니까 확률 떨어지기 시작..
  kn.n_neighbors=n
  score=kn.score(fish_data,fish_target)
  print(n,score)
```
